```python
import cv2
import matplotlib.pyplot as plt
import glob
import numpy as np
import random

path_load = glob.glob("/home/administrator/Desktop/Pham_Thai/Project/Resolution_CMND/Dataset/Data_Train/CMND_Standard/*.jpg")
path_save = "/home/administrator/Desktop/Pham_Thai/Project/Resolution_CMND/Dataset/Data_Train/CMND_Noise/"
dim = len(path_load)
print(dim)
i = 0
print("Running.......!!!!!!!")

def get_name(link):
    stri = ''
    link = link[::-1]
    d = len(link)
    for i in range(0, d):
        if link[i] == "/":
            break
        else:
            stri += link[i]
    stri = stri[::-1]
    return stri

while i < dim:
    a = random.randint(0,6)
    im = plt.imread(path_load[i])
    name = get_name(path_load[i])
    name = "{}".format(str(name))
    try :
        if a == 0:
            dst = cv2.GaussianBlur(im[0:50:],(15,15),cv2.BORDER_DEFAULT)
            cv2.imwrite(path_save + name, dst)
        elif a == 1:
            kernel = np.ones((5,10),np.float32)/50
            dst = cv2.filter2D(im,-1,kernel)
            cv2.imwrite(path_save + name, dst)
        elif a == 2:
            kernel = np.ones((3,3),np.float32)/25
            dst = cv2.filter2D(im,-1,kernel)
            cv2.imwrite(path_save + name, dst)
        elif a == 3:
            kernel = np.ones((2,10),np.float32)/25
            dst = cv2.filter2D(im,-1,kernel)
            cv2.imwrite(path_save + name, dst)
        elif a == 4:
            blur = cv2.blur(im,(2,10))
            cv2.imwrite(path_save + name, blur)
        elif a == 5:
            median = cv2.medianBlur(im,5)
            cv2.imwrite(path_save + name, median)
        elif a == 6:
            kernel = np.ones((1,10),np.float32)/25
            dst = cv2.filter2D(im,-1,kernel)
            cv2.imwrite(path_save + name, dst)
    except:
        print(path_save + name)
    i += 1
print(i)
print("Done........ !!!!!")
```
